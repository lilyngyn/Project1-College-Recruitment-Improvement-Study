```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('LNCLEANEDDATA.csv')

plt.figure(figsize=(15,5))
# What are the demographic characterists (gender, race/ethnicity, age) of students enrolled in these computing courses? 
# Question 1 for demographics(Gender, Race, and Age)
#Chart 1
plt.subplot(1,3,1)
df["gender"].value_counts().plot(kind="bar", color = "blue")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")

#Chart 2
plt.subplot(1,3,2)
df["race"].value_counts().plot(kind="bar", color = "red")
plt.title("Race/Ethnicity Distribution")
plt.xlabel("Race")
plt.ylabel("Count")

#Chart 3
plt.subplot(1,3,3)
df["age"].value_counts().plot(kind="bar", color = "red")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")

#Formatting
plt.tight_layout()
plt.show()
```

#### Answer
##### Gender Distribution: Women and men are the majority with women being the most, the minority is prefer not to say and Non-Binary people with Non-Binary people being the least
##### Race/Ethnicity Distribution: Most of the population consists of White students forming the largest group and then Hispanic or Latino students.
##### Age Distribution: Most students fall between 18-20 age range, with fewer people the older it gets.

```python
# Which degree programs are students enrolled in? 
# Question 2: Degree Programs

#Listing all the degrees in the dataset, combining them
print(df["degree_program"].unique())

degrees = {
    "Engineering": "Engineering",
    "Mechanical Engineering": "Engineering",
    "Electrical Engineering": "Engineering",
    "Civil Engineering": "Engineering",
    "Computer Science": "Computing & IT",
    "Comp Science": "Computing & IT",
    "Information Technology": "Computing & IT",
    "Cybersecurity": "Computing & IT",
    "Business Administration": "Business",
    "Finance": "Business",
    "Marketing": "Business",
    "Arts & Design": "Arts & Design",
    "Psychology": "Social Sciences",
    "Education": "Social Sciences",
    "Biology": "Sciences",
    "Chemistry": "Sciences",
    "Physics": "Sciences",
    "Other": "Other"
}

#Groups similar degrees together making it easier to put it into one category
df["degree_category"] = df["degree_program"].map(degrees).fillna("Other") 

plt.figure(figsize=(8, 5))
df["degree_category"].value_counts().plot(kind="bar", color="lightcoral")
plt.title("Degree Enrollment")
plt.xlabel("Degree")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# How did students hear about CCM?
# Question 3
#Gets the first 11 columns so I don't have to manually type out all columns
sources = df.iloc[:, :11] 
#Sums the amount of yes
counts = (sources == "Yes").sum()

plt.figure(figsize=(8, 8))
#Creating the pie chart
counts.plot(kind="pie", autopct="%1.1f%%", startangle = 140, colors=['darksalmon', 'lightgreen','teal' ,'coral' ,'lemonchiffon' ,'tomato', 'goldenrod' ,'dodgerblue','chocolate','lightblue','darkslategray' ])
plt.title("How Did Students Hear About CCM")
#hiding y label
plt.ylabel("")
plt.show()
```

#### Answer
##### Most people hear about CCM from friends or family members making that the biggest piece of advertising for enrollment for CCM.

```python
# How does interest in taking additional computing courses vary by demographic group (gender, age)?
# Question 4
plt.figure(figsize=(12,5))

#Creating the plots
plt.subplot(1,2,1)
df.groupby("gender")["interest_cs"].mean().plot(kind="bar", color = "blue")
plt.xticks(rotation=0)
plt.title("Interest in Computing Courses by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Interest Level")

plt.subplot(1,2,2)
df.groupby("age")["interest_cs"].mean().plot(kind="bar", color = "orange")
plt.xticks(rotation=0)
plt.title("Interest in Computing Courses by Age")
plt.xlabel("Age")
plt.ylabel("Average Interest Level")

plt.tight_layout()
plt.show()
```
